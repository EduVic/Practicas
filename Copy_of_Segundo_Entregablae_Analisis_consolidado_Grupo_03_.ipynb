```python
#'OTD1417'
alteration_limpio_pv1 = alteration_limpio.pivot_table(values='Intervalo_Alteration', index='HOLEID', columns='MALT_D', aggfunc='sum')
alteration_limpio_pv1.head()
```

```python
def analisis_alteration_holeid(tabla, holeid):
    filtro_holeid = tabla[tabla['HOLEID'] == holeid]

    if filtro_holeid.empty:
        return f"No se encontraron datos para HOLEID: {holeid}"

    geolfrom_min = filtro_holeid['GEOLFROM'].min()
    geolto_max = filtro_holeid['GEOLTO'].max()
    total_intervalo_alteration = filtro_holeid['Intervalo_Alteration'].sum()
    alteration_by_malt_d = filtro_holeid.groupby('MALT_D')['Intervalo_Alteration'].sum()

    results = {
        'HOLEID': holeid,
        'GEOLFROM_min': geolfrom_min,
        'GEOLTO_max': geolto_max,
        'Total_Intervalo_Alteration': total_intervalo_alteration,
        'Alteration_by_MALT_D': alteration_by_malt_d.to_dict()
    }

    return results

hole_id = 'OTD1417'
resultado_alteration = analisis_alteration_holeid(alteration_limpio, hole_id)
```

```python
#'OTD1417'
dry_bulk_density_limpio_pv1 = dry_bulk_density_limpio.pivot_table(values='Intervalo_DryBulkDensity', index='HOLEID', columns='BulkDensity', aggfunc='sum')
dry_bulk_density_limpio_pv1.head()
```

```python
def analisis_drybulkdensity_holeid(tabla1, holeid1):
    filtro_holeid1 = tabla1[tabla1['HOLEID'] == holeid1]

    if filtro_holeid1.empty:
        return f"No se encontraron datos para HOLEID: {holeid1}"

    geolfrom_mn = filtro_holeid1['GEOLFROM'].min()
    geolto_mx = filtro_holeid1['GEOLTO'].max()
    total_intervalo_drybulkdensity = filtro_holeid1['Intervalo_DryBulkDensity'].sum()
    drybulkdensity_by_bulkdensity = filtro_holeid1.groupby('BulkDensity')['Intervalo_DryBulkDensity'].sum()

    results = {
        'HOLEID': holeid1,
        'GEOLFROM_min': geolfrom_mn,
        'GEOLTO_max': geolto_mx,
        'Total_Intervalo_DryBulkDensity': total_intervalo_drybulkdensity,
        'drybulkdensity_BY_bulkdensity': drybulkdensity_by_bulkdensity.to_dict()
    }

    return results

hole_id1 = 'OTD1155'
resultado_drybulkdensity = analisis_drybulkdensity_holeid(dry_bulk_density_limpio, hole_id1)
```

Define the variables you want to plot. In this example the lithologies extracted from Leapfrog (assigned to the variable y1) and the ones logged by SRK (y1b) will be superposed on the same subplot/diagram for easier comparison:

```python
y = dataselected.loc[:, 'SAMPLEID']
y1 = dataselected.loc[:, 'SAMPFROM'].astype(float)
y1b = dataselected.loc[:, 'SAMPTO'].astype(float)
y2 = dataselected.loc[:, 'Au_ppm']
y3 = dataselected.loc[:, 'Ag_ppm']
y4 = dataselected.loc[:, 'Cu_pct']
y5 = dataselected.loc[:, 'SulphChalcopyrite']
y6 = dataselected.loc[:, 'SulphBornite']
```
