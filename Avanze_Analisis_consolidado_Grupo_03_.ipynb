```python
def analisis_alteration_holeid(tabla, holeid):
    filtro_holeid = tabla[tabla['HOLEID'] == holeid]

    if filtro_holeid.empty:
        return f"No se encontraron datos para HOLEID: {holeid}"

    geolfrom_min = filtro_holeid['GEOLFROM'].min()
    geolto_max = filtro_holeid['GEOLTO'].max()
    total_intervalo_alteration = filtro_holeid['Intervalo_Alteration'].sum()
    alteration_by_malt_d = filtro_holeid.groupby('MALT_D')['Intervalo_Alteration'].sum()

    results = {
        'HOLEID': holeid,
        'GEOLFROM_min': geolfrom_min,
        'GEOLTO_max': geolto_max,
        'Total_Intervalo_Alteration': total_intervalo_alteration,
        'Alteration_by_MALT_D': alteration_by_malt_d.to_dict()
    }

    return results

hole_id = 'OTD1417'
resultado_alteration = analisis_alteration_holeid(alteration_limpio, hole_id)
```

```python
def analisis_drybulkdensity_holeid(tabla1, holeid1):
    filtro_holeid1 = tabla1[tabla1['HOLEID'] == holeid1]

    if filtro_holeid1.empty:
        return f"No se encontraron datos para HOLEID: {holeid1}"

    geolfrom_mn = filtro_holeid1['GEOLFROM'].min()
    geolto_mx = filtro_holeid1['GEOLTO'].max()
    total_intervalo_drybulkdensity = filtro_holeid1['Intervalo_DryBulkDensity'].sum()
    drybulkdensity_by_bulkdensity = filtro_holeid1.groupby('BulkDensity')['Intervalo_DryBulkDensity'].sum()

    results = {
        'HOLEID': holeid1,
        'GEOLFROM_min': geolfrom_mn,
        'GEOLTO_max': geolto_mx,
        'Total_Intervalo_DryBulkDensity': total_intervalo_drybulkdensity,
        'drybulkdensity_BY_bulkdensity': drybulkdensity_by_bulkdensity.to_dict()
    }

    return results

hole_id1 = 'OTD1155'
resultado_drybulkdensity = analisis_drybulkdensity_holeid(dry_bulk_density_limpio, hole_id1)
```

```python
#import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Crear el widget de selección
holeid_dropdown = widgets.Dropdown(
    options=collar_limpio['HOLEID'].unique(),
    description='HOLEID:'
)

# Función para capturar la selección del usuario
def on_selection_change(change):
    global myHOLEID
    myHOLEID = change['new']
    is_present_in_alteration = myHOLEID in alteration_limpio['HOLEID'].values
    is_present_in_assay = myHOLEID in assay_limpio['HOLEID'].values
    is_present_in_dry_bulk_density = myHOLEID in alteration_dry_bulk_density['HOLEID'].values
    is_present_in_lithology = myHOLEID in lithology_limpio['HOLEID'].values
    is_present_in_mineralisation = myHOLEID in mineralisation_limpio['HOLEID'].values
    print(f'Selección guardada: {myHOLEID}')
    print(f'¿Está presente en alteration? {is_present_in_alteration}')
    print(f'¿Está presente en assay? {is_present_in_assay}')
    print(f'¿Está presente en dry_bulk_density? {is_present_in_dry_bulk_density}')
    print(f'¿Está presente en lithology? {is_present_in_lithology}')
    print(f'¿Está presente en mineralisation? {is_present_in_mineralisation}')

# Vincular la función al cambio en la selección
holeid_dropdown.observe(on_selection_change, names='value')

# Mostrar el combo de selección
display(holeid_dropdown)
```

### Concatenacion ***HOLEID (alteration - assay- mineralisation)***


```python
#UGD026
def concatenar(hole_id):
  ## Filtrando por hole_id
  alteration_limpio_OTD1417 = alteration_limpio[alteration_limpio['HOLEID'] == hole_id]
  assay_limpio_OTD1417 = assay_limpio[assay_limpio['HOLEID'] == hole_id]
  dry_bulk_density_limpio_OTD1417 = dry_bulk_density_limpio[dry_bulk_density_limpio['HOLEID'] == hole_id]
  lithology_limpio_OTD1417 = lithology_limpio[lithology_limpio['HOLEID'] == hole_id]
  mineralisation_limpio_OTD1417 = mineralisation_limpio[mineralisation_limpio['HOLEID'] == hole_id]
  collar_limpio_OTD1417 = collar_limpio[collar_limpio['HOLEID'] == hole_id]

  ## Concatenando las tablas
  alteration_mas_assay = pd.merge(alteration_limpio_OTD1417,assay_limpio_OTD1417,on='HOLEID',how='inner')
  #alteration_mas_assay_mas_dry = pd.merge(alteration_mas_assay,dry_bulk_density_limpio_OTD1417,on='HOLEID',how='inner')
  alteration_mas_assay_mas_dry = alteration_mas_assay
  alteration_mas_assay_mas_dry_mas_litho = pd.merge(alteration_mas_assay_mas_dry,lithology_limpio_OTD1417,on='HOLEID',how='inner')
  #alteration_mas_assay_mas_dry_mas_litho_mas_mineralization = pd.merge(alteration_mas_assay_mas_dry_mas_litho,mineralisation_limpio_OTD1417,on='HOLEID',how='inner')
  #alteration_mas_assay_mas_dry_mas_litho_mas_mineralization_mas_collar = pd.merge(alteration_mas_assay_mas_dry_mas_litho_mas_mineralization,collar_limpio_OTD1417,on='HOLEID',how='inner')
  alteration_mas_assay_mas_dry_mas_litho_mas_mineralization_mas_collar= alteration_mas_assay_mas_dry_mas_litho
  return alteration_mas_assay_mas_dry_mas_litho_mas_mineralization_mas_collar
```

# ***analisis respecto a la profundidad***

## ***Profundidad vs Lithologia***

```python
import plotly.graph_objects as go
import plotly.express as px

# Filtrar los datos para el pozo específico 'UGD026'
pozo_filtrado = lithology_limpio[lithology_limpio['HOLEID'] == 'UGD026']
unique_mlith = pozo_filtrado['mlith'].unique()

# Configurar el gráfico
fig = go.Figure()

# Iterar sobre cada tipo de litología y agregar barras apiladas
for mlith in unique_mlith:
    df_filtered = pozo_filtrado[pozo_filtrado['mlith'] == mlith]
    fig.add_trace(go.Bar(
        y=df_filtered['HOLEID'],
        x=df_filtered['to'] - df_filtered['from'],
        orientation='h',
        name=mlith
    ))
#gráfico
fig.update_layout(
    barmode='relative',
    title=('Litología del Pozo - UGD026'),
    xaxis_title='Profundidad',
    legend_title='Litología',
    legend=dict(
        x=1.02,
        y=1
    ),
    height=400,
    width=900
)
fig.show()
```
